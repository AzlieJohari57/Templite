### Data Transformation

````python
import json
from langchain import PromptTemplate, LLMChain
from langchain.schema import HumanMessage


# ---------------------------
# Helpers
# ---------------------------

def parse_llm_json_response(raw: str) -> list:
    """Sanitize and safely parse LLM output into JSON."""
    raw = raw.strip()
    if raw.startswith("```json"):
        raw = raw[len("```json"):].strip()
    elif raw.startswith("```"):
        raw = raw[len("```"):].strip()
    if raw.endswith("```"):
        raw = raw[:-3].strip()

    # Try extracting JSON array/object
    if not raw.startswith("[") and not raw.startswith("{"):
        json_start = raw.find("[") if "[" in raw else raw.find("{")
        json_end = raw.rfind("]") + 1 if "]" in raw else raw.rfind("}") + 1
        raw = raw[json_start:json_end]

    return json.loads(raw)


def build_chain(llm, template: str, input_variables: list):
    """Utility to create an LLMChain with a prompt template."""
    prompt = PromptTemplate(input_variables=input_variables, template=template)
    return LLMChain(llm=llm, prompt=prompt)


# ---------------------------
# Experience Enhancer
# ---------------------------

def _determine_targets(n_jobs: int):
    """Return target bullet count and word limit based on job count."""
    target = 8 if n_jobs == 1 else 4 if n_jobs == 2 else 3
    max_words = 45 if target in [3, 8] else 100
    return target, max_words


def enhance_experience(resume: dict, llm) -> dict:
    n_jobs = resume.get("number of jobs", 0)
    target, max_words = _determine_targets(n_jobs)

    template = """
    You are a professional resume writer.

    Company: {company}
    Title: {title}

    Existing bullets:
    {existing}

    Produce exactly {target} bullet points.

    RULES:
    - If more than {target} exist, keep the strongest, then rewrite them.
    - Each bullet must be clear, measurable, and professional.
    - Use industry language, highlight impacts with metrics.
    - Use context from {company}, {title}, {existing}.
    - Each bullet ≤ {max_words} words.

    Return ONLY a JSON array of strings.
    Example: ["Did X", "Improved Y", "Built Z"]
    """
    chain = build_chain(llm, template, ["company", "title", "existing", "target", "max_words"])

    for exp in resume.get("experience", []):
        existing_md = "\n".join(f"- {b}" for b in exp.get("details", []))
        try:
            raw = chain.run({
                "company": exp.get("company", ""),
                "title": exp.get("title", ""),
                "existing": existing_md,
                "target": target,
                "max_words": max_words
            })
            exp["details"] = parse_llm_json_response(raw)
        except Exception as e:
            print(f"⚠️ Failed to enhance {exp.get('title')}: {e}")
            continue

    return resume


# ---------------------------
# Strength Enhancer
# ---------------------------

def enhance_strength(resume: dict, llm) -> dict:
    n_jobs = resume.get("number of jobs", 0)

    if n_jobs == 0:
        template = """
        You are a resume builder AI.
        Candidate has no job experience.

        Generate 8 professional-strength bullet points (≤20 words each) for this role:
        Title: {title}

        Return only a JSON array of 8 strengths.
        """
    else:
        template = """
        You are a resume assistant.
        Summarize candidate's professional strengths in 1 item only (≤60 words), for this role:

        Title: {title}

        Return only a JSON array with one string.
        """

    chain = build_chain(llm, template, ["title"])
    try:
        raw = chain.run({"title": resume.get("title", "")})
        resume["strength"] = parse_llm_json_response(raw)
    except Exception as e:
        print(f"⚠️ Failed to parse strengths: {e}")
        resume["strength"] = []

    return resume


# ---------------------------
# Language Post-Processing
# ---------------------------

def enhance_resume_language(resume: dict, llm, language: str = "English") -> dict:
    """Apply final polishing & restructuring in selected language."""
    if language == "English":
        instructions = """
        You are a professional resume assistant.
        Enhance and structure resume JSON strictly in English.

        RULES:
        - Translate title to English.
        - Sort experiences latest → earliest.
        - Date format: DD/MM/YYYY.
        - All experiences/strengths must include measurable metrics.
        - Leave missing info blank (no fake data).
        - "about" must be polished, starting with level of experience, then elaborating.
        """
    else:
        instructions = """
        Anda ialah pembantu resume profesional.
        Struktur & perkemas resume dalam Bahasa Malaysia.

        PERATURAN:
        - Susun pengalaman terkini → terdahulu.
        - Format tarikh: DD/MM/YYYY.
        - Semua pengalaman/kekuatan mesti ada hasil boleh diukur.
        - Jika maklumat hilang, biarkan kosong.
        - "Tentang" mesti profesional, mula dengan tahap pengalaman, kemudian huraikan.
        """

    template = """
    Instructions:
    {instructions}

    Resume JSON:
    {resume_json}
    """
    chain = build_chain(llm, template, ["instructions", "resume_json"])

    try:
        raw = chain.run({
            "instructions": instructions,
            "resume_json": json.dumps(resume)
        })
        return parse_llm_json_response(raw)
    except Exception as e:
        print(f"⚠️ Failed language enhancement: {e}")
        return resume


# ---------------------------
# Pipeline Orchestration
# ---------------------------

def enhance_resume(resume: dict, llm, language="English") -> dict:
    resume = enhance_experience(resume, llm)
    resume = enhance_strength(resume, llm)
    resume = enhance_resume_language(resume, llm, language)
    return resume
````
